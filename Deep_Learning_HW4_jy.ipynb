### Linear regression

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import random
random.seed(10)
```

```python
regression = Sequential()
```

```python
regression.add(Dense(1, input_dim=2, activation='linear'))
```

```python
regression.compile(loss='mse', optimizer='sgd', metrics=['mse', 'mae'])
regress_result = regression.fit(X, y, epochs=20)
regression.evaluate(X,y)
```

### Support Vector Machine with Hinge loss

```python
import random
random.seed(13)
svm = Sequential()
svm.add(Dense(1, input_dim=2,activation='sigmoid', kernel_regularizer=tf.keras.regularizers.l2())) ##linear activation is by default, just add there for complete analysis
svm.compile(loss='hinge',
              optimizer='adadelta',
              metrics=['accuracy'])
random.seed(13)
svm_result = svm.fit(X, y, epochs=50)
svm.evaluate(X,y)
## observe an unstable training error pattern ranging from 20% to 75%
```

### Support vector regression

```python
## with regard to continuous outcome
## define epsilon-sensitive(ess)
import tensorflow.keras.backend as K
def svr_loss(layer, y_true, y_pred):
    weights = layer.weights[0]
    weights_tf = tf.convert_to_tensor(weights)
    y_true_0 = tf.cast(y_true, tf.float32)
    dif = abs(y_pred-y_true_0)
    ess_loss = K.sum(K.maximum(0.0,dif-0.1),axis=-1)## how to choose an optimal epsilon
    regularization_loss = 0.5*(tf.reduce_sum(tf.multiply(weights_tf,weights_tf)))
    return regularization_loss + 0.2*ess_loss ## how to choose an optimal penalization parameter
```

```python
## establish the svr framework
svr = Sequential()
svr.add(Dense(1, input_dim=2,activation='sigmoid',name='svrl'))
svr.compile(loss=lambda y_true,y_pred:svr_loss(svr.get_layer('svrl'),y_true,y_pred),optimizer='adam',metrics=['mse','acc'])
svr_result = svr.fit(X, y, epochs=50)
svr.evaluate(X,y)

```

## PartII: XOR function

```python
import tensorflow as tf
import numpy as np
```

```python
xor_x = np.array([[0,0],[0,1],[1,0],[1,1]])
xor_y = np.array([0,1,1,0])
xor = Sequential()
xor.add(Dense(units=15,activation='relu',input_dim=2))
xor.add(Dense(units=1,activation='sigmoid'))
xor.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
xor.fit(xor_x,xor_y,epochs=100,batch_size=4)
print(xor.predict(xor_x,batch_size=4))
## we map the value which is less than 0.5 to 0 and map the value which is more than 0.5 to 1
## Finally, we can get 100% accuracy to mimic xor function
```

From the prediction resultsm we can see that the neural network we constructed can mimic XOR function

## Part III: Simulate data and build neural network

```python
import scipy.stats
from sklearn.model_selection import train_test_split
df_x = np.random.standard_normal(size=(1000, 10))
chi_50 = scipy.stats.chi2.ppf(0.5, df=10)
print(chi_50)
sq = lambda vec_x:  sum([vec_x[i]**2 for i in range(0, 10)])
df_y = np.zeros(shape=1000)
for i in range(0, 1000):
    df_y[i] = 1 if sq(df_x[i]) > chi_50 else 0
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.33, random_state=42)
print(y_train[:10])
nn = Sequential()
nn.add(Dense(units=8,activation='relu',input_dim=10))
nn.add(Dense(units=10,activation='linear'))
nn.add(Dense(units=12,activation='linear'))
nn.add(Dense(units=1,activation='sigmoid'))
nn.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
nn.fit(x_train,y_train,epochs=50)
nn.evaluate(x_test,y_test)
## obtain accuracy of 0.8121 on test set
```
